#Libraries

```python
import pandas as pd
import scipy
from scipy import stats
```

#Data

```python
df = pd.read_csv("LabTAT.csv")
df
```

```python
Laboratory_1 = df['Laboratory 1']
Laboratory_2 = df['Laboratory 2']
Laboratory_3 = df['Laboratory 3']
Laboratory_4 = df['Laboratory 4']
```

#F-statisctiacl & P-value

```python
stats.f_oneway(Laboratory_1, Laboratory_2, Laboratory_3, Laboratory_4)
```

P-value > Alpha value i.e., 2.1156 > 0.05
Hence there is no siginicant difference in the average TAT among the laboratories
